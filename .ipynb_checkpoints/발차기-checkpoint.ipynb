```python
from collections import deque

def bfs(S, T):

    q = deque()
    q.append((S, T, 0))
    check = [-1]*(200) #빈 배열 만들기

    while q:
        node, t, c = q.popleft() #각각 s, t, 카운트
        if node <= t: # n <= t and check[node] == -1 아 그니까 -1이라는것은 해당 값에서 실행되지 않았을 때
            q.append((node*2, t+3, c+1))
            q.append((node+1, t, c+1))
            print(q)#결국에 현재 check[node]에는 +2가 되는 것이다.
            if node == t:
                print(c)
                return c
```

```python
C = int(input())
for _ in range(C):
    S, T = map(int, input().split())
    print(bfs(S, T))
```
