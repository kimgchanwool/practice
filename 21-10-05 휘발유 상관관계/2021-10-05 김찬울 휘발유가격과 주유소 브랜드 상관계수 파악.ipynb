```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/marketindex/oilDailyQuote.naver?marketindexCd=OIL_GSL&page='
num = list(map(str, list(range(93, 146))))
date_price = []
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

for i in num:
    res = requests.get(url+i, headers = headers)
    soup = BeautifulSoup(res.text, 'lxml')

    d = soup.find_all('td', attrs={'class' : 'date'})
    j = soup.find_all('td', attrs={'class' : 'num'})
    for i, k in zip(d, j[::3]):
        if i.text.strip().split('.')[0][2:] != '19':
            continue
        date = i.text.strip()
        date = ''.join(date.split('.'))
        a, c = k.text.split('.')
        a = int(''.join(a.split(',')))
        c = float(c) / 100
        k = a + c
        date_price.append([date, k])
```

```python
date_price = np.array(date_price)
```

```python
date_price = pd.DataFrame(date_price, columns=['기간', '실제가격'])
```
